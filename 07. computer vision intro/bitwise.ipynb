```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = np.zeros((500,500,3),np.uint8)
img1 = cv2.rectangle(img1,(40,100),(250,250),(255,255,255),-1)
plt.figure(figsize=(20,15))
plt.imshow(img1)
```

```python
img2 = np.zeros((500,500,3),np.uint8)
img2 = cv2.circle(img2,(250,250),100,(255,255,255),-1)
plt.figure(figsize=(20,15))
plt.imshow(img2)
```

```python
bitand = cv2.bitwise_and(img1,img2)
plt.figure(figsize=(8,8))
plt.imshow(bitand)
```

```python
bitor = cv2.bitwise_or(img1,img2)
plt.figure(figsize=(8,8))
plt.imshow(bitor)
```

```python
bitxor = cv2.bitwise_xor(img1,img2)
plt.figure(figsize=(8,8))
plt.imshow(bitxor)
```

```python
bitnot1 = cv2.bitwise_not(img1)
bitnot2 = cv2.bitwise_not(img2)

plt.figure(figsize=(8,8))
plt.imshow(bitnot1)

plt.figure(figsize=(8,8))
plt.imshow(bitnot2)
```
